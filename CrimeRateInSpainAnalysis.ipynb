# Data Cleaning

```python
crime_df.groupby("Crime")["Total cases"].sum().sort_values(ascending=False).plot.bar()
```

Other criminal offenses is not helpful in the data analysis so we are going to remove it

```python
crime_df = crime_df[crime_df.Crime!="Other criminal offenses"]
```

```python
crime_df.groupby("Crime")["Total cases"].sum().sort_values(ascending=False).plot.bar()
```

# Data selection

Select top three cities with the largest total crime

```python
crime_df.groupby("Location")["Total cases"].sum().sort_values(ascending=False).plot.bar()
```

# Methodology

Since the top three cities with the highest crime rates are Barcelona, Madrid, and Valencia, we will be using them to do the data analysis. Data will be analyzed based on the following questions:<br>
1)What are the top 5 crimes per city from 2019 to 2021?<br>
2)For each crime, how did it change from 2019 to 2021?

# Barcelona

```python
barcelona_df=crime_df[crime_df.Location == "Barcelona"]
```

```python
barcelona_df.groupby("Crime")["Total cases"].sum().sort_values().plot.barh()
```

The top five total crimes from 2019 to 2021 for Barcelona City are Theft, Robbery with violence and intimidation, Burglaries with force in homes, Robberies with force in homes, Vehicle Theft

```python
#selecting subsets based on the top 5 crimes
barcelona_crimes = barcelona_df.groupby("Crime")["Total cases"].sum().sort_values(ascending=False).index
barcelona_crimes1 = barcelona_df[barcelona_df.Crime == barcelona_crimes[0]].copy()
barcelona_crimes2 = barcelona_df[barcelona_df.Crime == barcelona_crimes[1]].copy()
barcelona_crimes3 = barcelona_df[barcelona_df.Crime == barcelona_crimes[2]].copy()
barcelona_crimes4 = barcelona_df[barcelona_df.Crime == barcelona_crimes[3]].copy()
barcelona_crimes5 = barcelona_df[barcelona_df.Crime == barcelona_crimes[4]].copy()

#set date time
barcelona_crimes1["Year"]=pd.to_datetime(barcelona_crimes1["Year"], format = "%Y")
barcelona_crimes2["Year"]=pd.to_datetime(barcelona_crimes2["Year"], format = "%Y")
barcelona_crimes3["Year"]=pd.to_datetime(barcelona_crimes3["Year"], format = "%Y")
barcelona_crimes4["Year"]=pd.to_datetime(barcelona_crimes4["Year"], format = "%Y")
barcelona_crimes5["Year"]=pd.to_datetime(barcelona_crimes5["Year"], format = "%Y")
```

```python
display(barcelona_crimes1)
display(barcelona_crimes2)
display(barcelona_crimes3)
display(barcelona_crimes4)
display(barcelona_crimes5)
```

```python
fig, ax = plt.subplots(ncols=5, figsize = (30,5))

sns.lineplot(data=barcelona_crimes1, x="Year", y="Total cases", ax=ax[0])
sns.lineplot(data=barcelona_crimes2, x="Year", y="Total cases", ax=ax[1])
sns.lineplot(data=barcelona_crimes3, x="Year", y="Total cases", ax=ax[2])
sns.lineplot(data=barcelona_crimes4, x="Year", y="Total cases", ax=ax[3])
sns.lineplot(data=barcelona_crimes5, x="Year", y="Total cases", ax=ax[4])

ax[0].set_title(barcelona_crimes[0])
ax[1].set_title(barcelona_crimes[1])
ax[2].set_title(barcelona_crimes[2])
ax[3].set_title(barcelona_crimes[3])
ax[4].set_title(barcelona_crimes[4])

ax[0].xaxis.set_major_locator(md.YearLocator())
ax[1].xaxis.set_major_locator(md.YearLocator())
ax[2].xaxis.set_major_locator(md.YearLocator())
ax[3].xaxis.set_major_locator(md.YearLocator())
ax[4].xaxis.set_major_locator(md.YearLocator())
```

It can be observed that for all top five highes crimes in Madrid, the highest total cases for each kind of crime was highest during 2019 and lowest during 2020. By 2021, total cases for theft, roberry with violence and intimidation, and burglaries with force in homes increased. However, for robberies with force in homes, total cases decreased and for vehicle theft, there was barely any change from 2020.

The top five total crimes from 2019 to 2021 for Madrid City are Theft, Robbery with violence and intimidation, Burglaries with force in homes, Robberies with force in homes, Vehicle Theft

It can be observed that for all top five highes crimes in Madrid, the highest total cases for each kind of crime was highest during 2019 and lowest during 2020 with a slow rise at 2021 which was not as high as 2019

# Valencia

```python
valencia_df=crime_df[crime_df.Location == "Valencia"]
```

```python
valencia_df.groupby("Crime")["Total cases"].sum().sort_values().plot.barh()
```

The top 5 crimes for Valencia city are theft, burglaries with force in homes, robberies with force in homes, robbery with violence and intimidation, and vehicle thet

```python
#selecting subsets based on top 5 crimes
valencia_crimes = valencia_df.groupby("Crime")["Total cases"].sum().sort_values(ascending=False).index
valencia_crimes1 = valencia_df[valencia_df.Crime == valencia_crimes[0]].copy()
valencia_crimes2 = valencia_df[valencia_df.Crime == valencia_crimes[1]].copy()
valencia_crimes3 = valencia_df[valencia_df.Crime == valencia_crimes[2]].copy()
valencia_crimes4 = valencia_df[valencia_df.Crime == valencia_crimes[3]].copy()
valencia_crimes5 = valencia_df[valencia_df.Crime == valencia_crimes[4]].copy()

valencia_crimes1["Year"] = pd.to_datetime(valencia_crimes1["Year"], format="%Y")
valencia_crimes2["Year"] = pd.to_datetime(valencia_crimes2["Year"], format="%Y")
valencia_crimes3["Year"] = pd.to_datetime(valencia_crimes3["Year"], format="%Y")
valencia_crimes4["Year"] = pd.to_datetime(valencia_crimes4["Year"], format="%Y")
valencia_crimes5["Year"] = pd.to_datetime(valencia_crimes5["Year"], format="%Y")
```

```python
display(valencia_crimes1)
display(valencia_crimes2)
display(valencia_crimes3)
display(valencia_crimes4)
display(valencia_crimes5)
```

```python
fig, ax = plt.subplots(ncols=5, figsize=(30,5))
sns.lineplot(data=valencia_crimes1,x="Year", y="Total cases", ax=ax[0])
sns.lineplot(data=valencia_crimes2,x="Year", y="Total cases", ax=ax[1])
sns.lineplot(data=valencia_crimes3,x="Year", y="Total cases", ax=ax[2])
sns.lineplot(data=valencia_crimes4,x="Year", y="Total cases", ax=ax[3])
sns.lineplot(data=valencia_crimes5,x="Year", y="Total cases", ax=ax[4])

ax[0].set_title(valencia_crimes[0])
ax[0].xaxis.set_major_formatter(md.DateFormatter('%Y'))
ax[0].xaxis.set_major_locator(md.YearLocator())

ax[1].set_title(valencia_crimes[1])
ax[1].xaxis.set_major_formatter(md.DateFormatter('%Y'))
ax[1].xaxis.set_major_locator(md.YearLocator())

ax[2].set_title(valencia_crimes[2])
ax[2].xaxis.set_major_formatter(md.DateFormatter('%Y'))
ax[2].xaxis.set_major_locator(md.YearLocator())

ax[3].set_title(valencia_crimes[3])
ax[3].xaxis.set_major_formatter(md.DateFormatter('%Y'))
ax[3].xaxis.set_major_locator(md.YearLocator())

ax[4].set_title(valencia_crimes[4])
ax[4].xaxis.set_major_formatter(md.DateFormatter('%Y'))
ax[4].xaxis.set_major_locator(md.YearLocator())
```

It can be observed that the highest total cases from 2019 to 2021 was highest at 2019 for four crimes: theft, burglaries with force in homes, robberies with force in homes, and vehicle theft. For robbery with violence and intimidation, the highest total cases was at 2021. For all 5 crimes, the lowest was at 2020 and for theft, burglaries with force in homes, robberies with force in homes, and robbery with violence and intimidation, cases rapidly icnreased close to 2019 level by 2021. For vehicle theft, there was a small increase from 2020 to 2021.

# Conclusion

Barcelona, Madrid, and Valencia had the same top 5 crimes with the highest total cases from 2019 to 2021. For all three cities, 2019 had the highest total cases and 2020 had the lowest total cases for all top 5 crimes. For Barcelona and Madrid, there was a small increase in crime from 2020 to 2021 but in Valencia, there was a high increase that almost reached the 2019 level for most of the crimes.

The decrease in crime for 2020 can be caused by the pandemic where there was less crime because people were getting sick and there were stricter government protocols at that time.
